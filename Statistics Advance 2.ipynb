 Q.1: : What is hypothesis testing in statistics?


   ->Hypothesis testing is a formal procedure for making decisions about a population parameter based on sample data. You state two competing hypotheses (null and alternative), choose a test statistic and significance level, compute the statistic from your sample, and decide whether the evidence is strong enough to reject the null hypothesis in favor of the alternative.

 Q.2: What is the null hypothesis, and how does it differ from the alternative
hypothesis?


  ->Null hypothesis (H₀): the default claim (e.g., population mean μ = μ₀). It represents no effect or no difference.

Alternative hypothesis (H₁ or Hₐ): what you suspect might be true instead (e.g., μ ≠ μ₀, μ > μ₀, or μ < μ₀).
You test whether sample evidence contradicts H₀ strongly enough to prefer H₁.

 Q.3 Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.



  ->The significance level α (commonly 0.05) is the threshold probability of making a Type I error (rejecting H₀ when it is true). If the p-value ≤ α you reject H₀; otherwise you fail to reject H₀. α controls how strict the test is — smaller α makes it harder to reject H₀.

 Q.4: What are Type I and Type II errors? Give examples of each.


  -> Type I error (false positive): rejecting H₀ when H₀ is true. Example: Concluding a new drug works when it actually doesn’t.

Type II error (false negative): failing to reject H₀ when H₁ is true. Example: Failing to detect a real effect of a drug (concluding it doesn’t work when it does).
The probability of Type I error = α. The probability of Type II error is denoted β; test power = 1 − β.

 Q.5: What is the difference between a Z-test and a T-test? Explain when to use
each.


  ->Z-test: used when the population standard deviation σ is known (or when sample size is large enough to approximate σ well). The test statistic uses the normal (Z) distribution.

T-test: used when σ is unknown and the sample size is small; it uses the Student’s t-distribution with n−1 degrees of freedom. For large n (say n > 30), t and z give similar results.

 Q.6 Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters of binomial distribution
n = 10        # number of trials
p = 0.5       # probability of success
size = 10000  # number of samples

# Generate binomial distribution data
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=11, edgecolor='black')
plt.title("Histogram of Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()

# Calculate mean and variance
print("Mean =", np.mean(data))
print("Variance =", np.var(data))
```

 Q.7 : Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

```python
import numpy as np
from scipy import stats

# Sample dataset
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

#Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}

# Hypothesized population mean
mu = 50

# Sample statistics
n = len(sample_data)
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)

# Z-statistic
z = (sample_mean - mu) / (sample_std / np.sqrt(n))

# p-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("Z-statistic:", z)
print("P-value:", p_value)
```

Sample Mean ≈ 50.08
Sample Standard Deviation ≈ 0.55
Z-Statistic ≈ 0.87
P-value ≈ 0.38

Interpretation of Results:
Null Hypothesis 𝐻0: The population mean is 50.
Alternative Hypothesis 𝐻1: The population mean is not equal to 50.

Since:
p-value (0.38) > 0.05
we fail to reject the null hypothesis.

Conclusion
There is no significant evidence to suggest that the sample mean differs from the population mean of 50.
The sample data appears to be consistent with the assumed population mean.

 Q.8 Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from a normal distribution
np.random.seed(0)
data = np.random.normal(loc=50, scale=5, size=100)

# Step 2: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

# Step 3: Calculate 95% confidence interval
confidence = 0.95
z_value = stats.norm.ppf((1 + confidence) / 2)

margin_error = z_value * (sample_std / np.sqrt(n))

lower_bound = sample_mean - margin_error
upper_bound = sample_mean + margin_error

print("Sample Mean:", sample_mean)
print("95% Confidence Interval:", (lower_bound, upper_bound))

# Step 4: Plot the histogram
plt.hist(data, bins=15, edgecolor='black')
plt.title("Histogram of Simulated Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

Interpretation
The data is generated from a normal distribution with mean 50 and standard deviation 5.
The 95% confidence interval estimates the range in which the true population mean is likely to fall.
This means we are 95% confident that the true mean of the population lies between the calculated lower and upper bounds.
The histogram visually shows the bell-shaped distribution of the simulated data.

 Q.9 Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

def analyze_data_with_zscores(data):
    """
    Calculates Z-scores for a given dataset and visualizes the standardized data
    using a histogram.

    Args:
        data (array-like): The input dataset.
    """
    # Calculate Z-scores
    z_scores = zscore(data)

    # Visualize standardized data with a histogram
    plt.figure(figsize=(8, 6))
    plt.hist(z_scores, bins=20, edgecolor='black')
    plt.title('Histogram of Standardized Data (Z-scores)')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

    return z_scores

# Example usage:
sample_data = np.array([10, 12, 15, 12, 18, 20, 22, 11, 14, 16, 30, 8])
calculated_z_scores = analyze_data_with_zscores(sample_data)
print("Calculated Z-scores:")
print(calculated_z_scores)
```
